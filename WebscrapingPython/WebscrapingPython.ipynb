```python
def parse_elements(url,spans):
    session = HTMLSession()
    r = session.get(url)
    spans = r.html.find(spans)
    return spans
```

```python
spans = parse_elements(url, 'span')
# links = parse_elements(url, 'a')
print(spans)
```

```python
def parse_href(url, element):
    session = HTMLSession()
    r = session.get(url)
    elements = r.html.find(element)
    links = [element.attrs['href'] for element in elements if 'href' in element.attrs]
    return links
```
